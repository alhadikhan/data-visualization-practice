#Matplotlib

```python
from matplotlib import pylab
print(pylab.__version__)
```

```python
import numpy as np
x=np.linspace(0,10,25)
y=x*x+2
print(x,y)
```

```python
print(np.array([x,y]).reshape(25,2).reshape(2,25))
```

```python
pylab.plot(x,y,'g')
```

```python
pylab.subplot(1,2,1) #rows,columns,indexes
pylab.plot(x,y,'g--') #3rd has determines color and line style
pylab.subplot(1,2,2)
pylab.plot(y,x,'r--')
```

```python
from matplotlib import pyplot as plt
```

```python
fig=plt.figure()
axis=fig.add_axes([0.5,0.1,0.8,.2]) # controls left,right,width,height of the canvas from (0,1)
axis.plot(x,y,'g--')
fig=plt.figure()
axis=fig.add_axes([0.5,0.1,0.8,2]) # controls left,right,width,height of the canvas from (0,1)
axis.plot(y,x,'r--')
```

```python
#we can draw subgraphs
fig, axes = plt.subplots(nrows=2, ncols=1) #submap is of 1 row, 2 columns
for ax in axes:
    ax.plot(x,y,'g')
```
