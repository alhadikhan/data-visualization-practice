```python
from matplotlib import pyplot as plt
```

```python
#add_axes[leftright,updown,width,hight]
fig=plt.figure()
axis=fig.add_axes([0.5,0.1,0.8,.8]) # controls left,right,width,height of the canvas from (0,1)
axis.plot(x,y,'g--')
axis=fig.add_axes([0.2,0.5,0.2,.2]) # controls left,right,width,height of the canvas from (0,1)
axis.plot(y,x,'r--')
```

```python
fig=plt.figure()
axis1=fig.add_axes([0.1,0.1,0.8,0.8]) # controls left,right,width,height of the canvas from (0,1)
axis2=fig.add_axes([0.2,0.5,0.3,0.5])
axis1.plot(x,y,'g--') # controls left,right,width,height of the canvas from (0,1)
axis2.plot(y,x,'r--')
```

```python
#we can draw subgraphs
fig, axes = plt.subplots(nrows=2, ncols=1) #submap is of 1 row, 2 columns
for ax in axes:
    ax.plot(x,y,'g')
```

```python
fig= plt.figure(figsize=(16,9),dpi=100) #New graphic object
fig.add_subplot()
plt.plot(x,y,'g')
```

```python
fig,axes=plt.subplots(figsize=(16,9),dpi=50)
axes.plot(x,y,'g')
```

```python
fig,axes=plt.subplots()
axes.set_xlabel("x")
axes.set_ylabel("y")
axes.set_title("title")
axes.plot(x,y,'g')
```

```python
ax.legend(["label1","label2"])
fig,axes=plt.subplots()
axes.set_xlabel("x")
axes.set_ylabel("y")
axes.set_title("title")
axes.plot(x,x**3)
axes.plot(x,x**2.5)
axes.legend(["y=x**3","y=x**2.5"],loc=2)
```

```python
#in matplotlib we can set properties such as line color,transparency and more
fig,axes=plt.subplots(dpi=100)
axes.plot(x,x+1,color="purple")
axes.plot(x,x+2,color="pink")
axes.plot(x,x+3,color="black")
#x,x+1,color="green"
```

```python
fig,axes=plt.subplots(dpi=100)
axes.plot(x,x**2,color="purple")
axes.plot(x,x+2,color="green",alpha=.5)
axes.plot(x,x+3,color="black",alpha=.5)
```

```python
fig,axes=plt.subplots(dpi=100)
axes.plot(x,x+1,color="purple",linewidth=5)
axes.plot(x,x+2,color="green",linewidth=2)
axes.plot(x,x+3,color="black",linewidth=.5)
```

```python
fig,axes=plt.subplots(dpi=100)
axes.plot(x,x+1,color="green",lw=2,linestyle='-')
axes.plot(x,x+2,color="green",lw=2,linestyle='-.')
axes.plot(x,x+3,color="green",lw=2,linestyle=':')
line, =axes.plot(x,x+4,color="black",lw=1.5)
line.set_dashes([5,10,15,10])
line, =axes.plot(x,x+5,color="black",lw=1.5)
line.set_dashes([5,10,5,10])
line, =axes.plot(x,x+6,color="black",lw=1.5)
line.set_dashes([5,10,15,20])
```
